# Partie 1 : Nettoyage des données

```python
import pandas as pd
import numpy as np

raw = pd.read_csv('SBAnational.csv')
raw.columns
```

```python
raw[['ApprovalDate','DisbursementDate','Term','ChgOffDate']].dropna()
```

```python
raw.columns
```

```python
raw = raw[raw.NewExist!=0]
```

```python
df = raw[['State','NAICS','Term','GrAppv','MIS_Status','NoEmp', 'NewExist','CreateJob', 'FranchiseCode', 'UrbanRural']]
df['GrAppv'] = df['GrAppv'].str.replace(r'$','')
df['GrAppv'] = df['GrAppv'].str.replace(r',','').astype(float)
# df.DisbursementGross = df.DisbursementGross.str.replace(r'$','')
# df.DisbursementGross = df.DisbursementGross.str.replace(r',','').astype(float)

df


```

```python
df = df[df.Term!=0]
```

```python
df.NAICS=df.NAICS.astype('string')
df.NAICS = df.NAICS.str[:2]
```

```python
df.info()
```

```python
liste = ['0','11','21','22','23','31','32','33','42','44','45','48','49','51','52','53','54','55','56','61','62','71','72','81']
```

```python
liste2 = ['Others','Agriculture, forestry, fishing and hunting','Mining, quarrying, and oil and gas extraction','Utilities','Construction','Manufacturing','Manufacturing','Manufacturing','Wholesale trade','Retail trade','Retail trade','Transportation and warehousing','Transportation and warehousing','Information','Finance and insurance','Real estate and rental and leasing','Professional, scientific, and technical services','Management of companies and enterprises','Administrative and support and wase management and remediation services','Educational services','Health care and social assistance','Arts, entertainment, and recreation','Accommodation and food services','Other services (except public administration) 92 Public administration']
```

```python
dict = {}
for key in liste:
    for value in liste2:
        dict[key] = value
        liste2.remove(value)
        break
dict
```

```python
df.NAICS = df.NAICS.map(dict)
df.MIS_Status = df.MIS_Status.map({'P I F': '0','CHGOFF':'1'})
```

```python
df = df[df['State'].notna()]
```

```python
df.dropna(inplace=True)
df.isnull().sum()
```

```python
df.State = df.State.astype('string')
df.NAICS = df.NAICS.astype('string')
df.MIS_Status = df.MIS_Status.astype(int)
```

```python
df.columns = ['State', 'Industry', 'Term', 'GrAppv', 'MIS_Status','NoEmp', 'NewExist', 'CreateJob', 'RetainedJob', 'FranchiseCode','UrbanRural']
```

```python
df
```

```python
df['Real_estate'] = df.Term.apply(lambda x: 0 if x<240 else 1 )
```

```python
df.NewExist = df.NewExist.apply(lambda x: 0 if x == 1 else 1)
df.NewExist.value_counts()
df.isnull().sum()
df.FranchiseCode=df.FranchiseCode.apply(lambda x: 1 if x ==0 else x)
df.FranchiseCode.value_counts()
```

```python
df.NewExist = df.NewExist.astype(str)
df.UrbanRural= df.UrbanRural.astype(str)
df.Real_estate = df.Real_estate.astype(str)
df.MIS_Status = df.MIS_Status.astype(str)
df.FranchiseCode = df.FranchiseCode.astype(str)
```

```python
df
```

```python
df.to_csv('data.csv',index=False)
```
